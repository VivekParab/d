```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
sb.get_dataset_names()
```

```python
df1 = pd.read_csv('forestfires.csv')
df2 = pd.read_csv('AirQuality.csv', sep=';')
df3 = pd.read_csv('adult.csv')
```

## Countplot (forest-fire)

```python
plt.title('Countplot Gender - Maritalstatus')
sb.countplot(data=df3, x='gender', palette='pastel', hue='marital-status')
plt.show()
```

```python
plt.figure(figsize=(8,7))
plt.title('Countplot Gender - Workclass')
sb.countplot(data=df3, x='gender', hue='workclass')
plt.show
```

## Histplot (Forest fire)

```python
plt.figure(figsize=(8, 7))
sb.histplot(data=df1, x='month', hue='wind', palette='pastel')
plt.title('Histplot Temperature') 
plt.show()
```
